```
def choose():
  choice="a"
  while choice not in ["0","1","2","3"]:
    choice=input("Choose between 0 to 3: ")
  if choice not in ["0","1","2","3"]:
    print("Choose between 0 to 3:")
  else:
    return int(choice)
```

```
choose()
```
